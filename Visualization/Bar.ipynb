### 柱状图绘制

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

安全情况柱状图：为了实现柱状图的横向对比，使用subplot制造画布，画轴

```python
# 数据准备
degree = ["安全", "较安全", "一般安全", "较不安全", "不安全"]
nanhu = [0.107, 0.473, 0.244, 0.080, 0.077]
jianhu = [0.092, 0.405, 0.345, 0.098, 0.060]
dongyuan = [0.006, 0.063, 0.307, 0.479, 0.134] 
xiyuan = [0.024, 0.366, 0.423, 0.188, 0.000]

fig, ax = plt.subplots(figsize=(10, 5))
xticks = np.arange(len(degree))
ax.bar(xticks, nanhu, width=0.16, label="南湖", color="white", edgecolor='black', hatch='---')
ax.bar(xticks+0.18, jianhu, width=0.16, label="鉴湖", color="white", edgecolor='black', hatch='')
ax.bar(xticks+0.36, dongyuan, width=0.16, label="东院", color="white", edgecolor='black', hatch='//')
ax.bar(xticks+0.54, xiyuan, width=0.16, label="西院", color="white", edgecolor='black', hatch='**')
ax.set_xlabel("安全评级")
ax.set_ylabel("隶属度")
plt.rcParams['font.sans-serif'] = ['SimHei']  # 替换sans-serif字体
plt.rcParams['axes.unicode_minus'] = False  # 解决坐标轴负数的负号显示问题
ax.legend()
ax.set_xticks(xticks+0.25)
ax.set_xticklabels(degree)
```

年龄柱状图，引入了seaborn中的蓝色背景，进行横向柱状图的对比

```python
# 背景调整
import seaborn as sns
sns.set()

# 数据准备
ages = ["18岁及以下", "19-25岁", "26-40岁", "41-55岁", "55岁及以上"]
yixian = [7, 61, 145, 60, 13]
xinyixian = [7, 82, 188, 83, 41]
shenghui = [4, 33, 61, 27, 13]

fig, ax = plt.subplots(figsize=(10, 5))
xticks = np.arange(len(ages))
ax.bar(xticks, yixian, width=0.25, label="一线城市", color='#99b8d7')
ax.bar(xticks+0.26, xinyixian, width=0.25, label="新一线城市", color='#b3b3b3')
ax.bar(xticks+0.52, shenghui, width=0.25, label="省会城市", color='#fbc740')
ax.set_xlabel("年龄")
ax.set_ylabel("人数")
plt.rcParams['font.sans-serif'] = ['SimHei']  # 替换sans-serif字体
plt.rcParams['axes.unicode_minus'] = False  # 解决坐标轴负数的负号显示问题
ax.legend()
ax.set_xticks(xticks+0.26)
ax.set_xticklabels(ages)
x_yixian = ax.get_xticks()-0.26
x_xinyixian = ax.get_xticks()
x_shenghui = ax.get_xticks()+0.26
x = [x_yixian, x_xinyixian, x_shenghui]
y = [yixian, xinyixian, shenghui]
for group in range(len(x)):
    for i in range(len(y[group])):
        xy = (x[group][i], y[group][i]+0.2)
        s = str(y[group][i])
        ax.annotate(
            s, # 要添加的文本
            xy = xy, # 位置
            fontsize=12, # 标签大小
            color='#404040', # 标签颜色
            ha = "center", # 水平对齐
            va = "baseline" # 垂直对齐
        )
```

```python
import seaborn as sns
sns.set()

# 数据准备
cities = ["一线城市", "新一线城市", "省会城市"]  # 作为标签
colors = ["#99b8d7", "#b3b3b3", "#fbc740"]
income = ["2500元及以下", "(2500,5000]元", "(5000,8000]元", "(8000,17000]元", "(17000,30000]元", "30000元以上"]
number1 = [36, 93, 55, 45, 23, 34]
number2 = [61, 134, 73, 71, 33, 29]
number3 = [21, 49, 28, 23, 5, 12]

# 图形和轴  #b3b3b3   #fbc740
fig, ax = plt.subplots(1, figsize=(10, 5))
ax.bar(income, number1, color="#99b8d7", label="一线城市")
ax.bar(income, number2, color="#b3b3b3", bottom=number1, label="新一线城市")
ax.bar(income, number3, color="#fbc740", bottom=np.array(number1)+np.array(number2), label="省会城市")
ax.set_xlabel("月收入")
ax.set_ylabel("人数")
ax.legend()
plt.rcParams['font.sans-serif'] = ['SimHei']  # 替换sans-serif字体
plt.rcParams['axes.unicode_minus'] = False  # 解决坐标轴负数的负号显示问题
x = ax.get_xticks()
start = [[0, 0, 0, 0, 0, 0],
         [36, 93, 55, 45, 23, 34],
         [97, 227, 128, 116, 56, 63]
        ]
y1 = number1
y2 = list(np.array(y1)+np.array(number2))
y3 = list(np.array(y2)+np.array(number3))
numbers = [number1, number2, number3]
ys = [y1, y2, y3]
for i in range(len(x)):
    for j in range(len(ys)):
        xy = (x[i], (ys[j][i]+start[j][i])/2)
        text = str(numbers[j][i])
        ax.annotate(
            text,
            xy = xy,
            fontsize=12,
            color='#404040',
            ha="center",
            va="baseline"
        )
```
